```python
import selenium
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import os

idioma='español'
try:
    os.mkdir(os.getcwd() + f'\\descargas\\{idioma}')
except:
    pass

download_file_path = os.getcwd() + f'\\descargas\\{idioma}'


def setting_selenium_options(download_file_path:str)->webdriver.ChromeOptions:
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-extensions")
    options.add_argument("--proxy-server='direct://'")
    options.add_argument("--proxy-bypass-list=*")
    options.add_argument("--start-maximized")
    options.add_argument('--disable-gpu')
    #options.add_argument('--headless')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--no-sandbox')
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--allow-running-insecure-content')
    user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.50 Safari/537.36'
    options.add_argument(f'user-agent={user_agent}')
    prefs = {
        "download.default_directory":download_file_path
        #"download.prompt_for_download": False,
        #"download.directory_upgrade": True
        }

    options.add_experimental_option('prefs', prefs)
    return options

```

```python
def Login(driver):

    user='cesar.bustossanchez@teleperformance.com'
    passw='Password001**'
    
    driver.get('chrome://settings/')
    driver.execute_script('chrome.settingsPrivate.setDefaultZoom(0.8);')
    if idioma == "español":
        driver.get('https://tpcmesp-co.callminer.net/Dashboard')
    if idioma =="ingles":
        driver.get('https://tpcmsite500.callminer.net/Search/Query')

    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="authtype-1"]'))).click()
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="submitButton"]'))).click()
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="Email"]'))).send_keys(user)
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="Password"]'))).send_keys(passw)
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="signInBtn"]'))).click()
```

```python
import pandas as pd
import os
import time

metadata = pd.read_excel(f"C:\Clover\metadata\metadata_{idioma}.xlsx", header=4)
links=metadata["Contact Link"].tolist()

def download():
    options = setting_selenium_options(download_file_path = download_file_path)
    driver = webdriver.Chrome("chromedriver.exe", options=options)
    
    Login(driver)

    for link in links:
        driver.get(link)
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH,'//*[@title="Download the transcript"]'))).click()
        time.sleep(1.5)

        

    
```

```python
eureka_id = []
texts=[]
database ={}

for i in os.listdir(f"descargas\\{idioma}"):
    with open (f'descargas\\{idioma}\\'+i, encoding='utf-8') as f:
        text = f.read()
    text = text.replace('\n','')
    text = text.replace('\t','')
    eureka_id.append(i.split('.')[0])
    texts.append(text)


database['Eureka ID'] = eureka_id
database['Feedback'] = texts



df = pd.DataFrame.from_dict(database, orient='columns')
df['Eureka ID'] = df['Eureka ID'].astype(int)








```
